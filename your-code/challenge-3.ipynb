## Q2: How to identify which country has the most VIP Customers?

```python
# counting total amout spent
total_amount_spent = orders.groupby('CustomerID')['amount_spent'].sum().reset_index()
print(total_amount_spent)
```

```python
# I select for my VIP Customers top 10% of all customers
```

```python
vip_customers = orders.groupby('CustomerID')['amount_spent'].quantile(0.10).reset_index()
vip_customers
```

```python
tenpercentile = vip_customers['amount_spent'].quantile(0.90)
print(tenpercentile)
```

```python
top_customers = total_amount_spent[total_amount_spent['amount_spent'] > tenpercentile]
print(top_customers)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
grouped_countries = orders.groupby(['CustomerID', 'Country'])['amount_spent'].sum().reset_index()
grouped_countries
```

```python
highest_spending_row = grouped_countries.loc[grouped_countries['amount_spent'].idxmax()]
highest_spending_row
```
